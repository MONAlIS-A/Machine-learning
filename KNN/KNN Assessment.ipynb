```python
import pandas as pd
import warnings as wr
wr.filterwarnings('ignore')
```

```python
df = pd.read_csv('weight-height.csv')
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df.Gender = le.fit_transform(df[['Gender']])
```

```python
df.head()
```

```python
df.tail()
```

#  Least Squared Linear Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# prepare features and target

X=df[['Gender','Height']]
y=df['Weight']
# split the data
X_train, X_test, y_train, y_test=train_test_split(X ,y, test_size=0.30 ,random_state=42)

# train the model 

model=LinearRegression()
model.fit(X_train, y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python

# Make predictions

y_pred=model.predict(X_test)

print(y_pred)

# Evaluate the model

mse=mean_squared_error(y_test, y_pred)
r2= r2_score(y_test, y_pred)

print(f'Mean Squared Error:{mse}')
print(f'R-squared :{r2}')



```

# KNN  Regressor Problem 

```python
from sklearn.neighbors import KNeighborsRegressor
# train the kNN Regressor
knn = KNeighborsRegressor(n_neighbors=100)
x=df[['Gender','Height']]
y2=df['Weight']
x_train, x_test, y2_train, y2_test=train_test_split(x ,y2, test_size=0.30 ,random_state=42)
knn.fit(x_train, y2_train)
```

```python
# make prdictions 
y_pred=knn.predict(x_test)
print(y_pred)
```

```python
# Evaluate the Model
mse= mean_squared_error(y2_test,y_pred)
print(f'Mean Squared Error:{mse}')

r2=r2_score(y_test, y_pred)
print(f'R-squared :{r2}')
```

# KNN Classification Problem

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

classifier=KNeighborsClassifier(n_neighbors=100)
```

```python
# prepare features and lables
x_=df[['Weight','Height']]
y_=df['Gender']
# split the dataset 
x__train, x__test, y__train, y__test = train_test_split(x_, y_, test_size=0.30, random_state=42)
```

```python
classifier.fit(x__train, y__train)
# make predictions
y1_pred=classifier.predict(x__test)
print(y1_pred)
```

```python
accuracy=classifier.score(x__test, y__test)
print(f'Accuracy: {accuracy * 100:.2f}%')
```
